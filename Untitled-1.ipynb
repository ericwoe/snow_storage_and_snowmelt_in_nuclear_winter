```python
def diagnose_time_coordinate(ds, label="Dataset"):
    """Analysiert die Zeit-Koordinate eines Datasets"""
    print(f"\n{'='*60}")
    print(f"{label}")
    print(f"{'='*60}")
    
    # Basis-Infos
    print(f"time dtype: {ds.time.dtype}")
    print(f"time shape: {ds.time.shape}")
    
    # Erste und letzte Werte
    print(f"\nFirst time value: {ds.time.values[0]}")
    print(f"  Type: {type(ds.time.values[0])}")
    print(f"Last time value: {ds.time.values[-1]}")
    print(f"  Type: {type(ds.time.values[-1])}")
    
    # Prüfe auf gemischte Typen
    time_types = [type(t).__name__ for t in ds.time.values]
    unique_types = set(time_types)
    
    print(f"\nUnique types in time coordinate: {unique_types}")
    print(f"Number of unique types: {len(unique_types)}")
    
    if len(unique_types) > 1:
        print("\n⚠️  WARNING: MIXED TYPES DETECTED!")
        for utype in unique_types:
            count = time_types.count(utype)
            print(f"  - {utype}: {count} values")
        
        # Finde erste Stelle wo Typ wechselt
        for i in range(len(time_types)-1):
            if time_types[i] != time_types[i+1]:
                print(f"\n  First type change at index {i}:")
                print(f"    [{i}] = {ds.time.values[i]} (type: {time_types[i]})")
                print(f"    [{i+1}] = {ds.time.values[i+1]} (type: {time_types[i+1]})")
                break
    else:
        print("✓ All time values have the same type")
    
    # Encoding und Attribute
    print(f"\ntime.encoding: {ds.time.encoding}")
    print(f"time.attrs: {ds.time.attrs}")
    
    # Prüfe ob np.diff funktionieren würde
    print("\nTesting np.diff compatibility:")
    try:
        diff_result = np.diff(ds.time.values[:3])
        print(f"  ✓ np.diff works: {diff_result}")
    except Exception as e:
        print(f"  ✗ np.diff FAILS: {e}")
    
    print(f"{'='*60}\n")
```

```python
import xarray as xr
```

```python
snow = xr.open_dataset("./results/snow_model_output.nc", engine="netcdf4")
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors as mcolors


data = snow["snow_storage_mm"].max(dim="time")
data = data/10
print(data)

bin_edges = np.array([0, 10, 100, 200, 300, 400, 600, 1000, data.max()])

# Anzahl Klassen
n_classes = len(bin_edges) - 1

# Diskrete Colormap erzeugen
#cmap = plt.get_cmap("YlGnBu", n_classes)
colors = [
    #"#006400", #0 - 0.1
    "#ffffff",  # 0–100        weiß
    "#e0f3f8",  # 100–1 000    sehr hellblau
    "#abd9e9",  # 1 000–2 000  hellblau
    "#74add1",  # 2 000–3 000  mittelblau
    "#4575b4",  # 3 000–4 000  dunkelblau
    "#313695",  # 4 000–6 000  sehr dunkelblau
    "#542788",  # 6 000–10 000 blau-lila
    "#3f007d",  # >10 000      lila
]

cmap = mcolors.ListedColormap(colors)

# Norm für Klassengrenzen
norm = mcolors.BoundaryNorm(bin_edges, ncolors=n_classes)

fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection=ccrs.Robinson())

im = data.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap=cmap,
    norm=norm,
    add_colorbar=True,
    cbar_kwargs={
        "label": "Snow storage [cm]",
        "ticks": bin_edges  # letzte (inf) nicht anzeigen
    }
)

ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.set_title("Maximum snow storage (classified)")

plt.show()

#Darstellung verbessern evtl. grün als Farbe, wo nie Schnee liegt
```
